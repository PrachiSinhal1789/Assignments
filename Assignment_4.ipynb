17. How do you distinguish between copy.copy() and copy.deepcopy()?

Ans:copy.copy() creates a new object which points or references towards the original object and changes made in new object is reflected in original object (valid for compound object i.e object containing object)
copy.deepcopy() creates a new object and adds elements into it from the original object and changes in new object is not reflected in original object
